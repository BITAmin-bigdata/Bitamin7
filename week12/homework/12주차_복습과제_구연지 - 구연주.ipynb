```python
menu_new = []
for menu in menus:
    menu_new.append(menu.split('\n'))

print(menu_new)
```

```python
menu1 = []
for element in menu_new:
    for j in element:
        menu1.append(j)

count = {}        
for i in menu1:   
    try: count[i] += 1
    except: count[i]=1
print(count)
```

```python
import pandas as pd
del count['[중식]']
del count['']
del count['[석식]']
menu_df = pd.DataFrame(list(count.items()),columns = ['menu','counts'])
menu_df.head()
```
