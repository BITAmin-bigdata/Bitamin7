```python
mean=[]
for i in [ratings_1,ratings_2,ratings_3,ratings_4,ratings_5,ratings_6]:
    n=sum(i)/len(i)
    mean.append(n)
mean
```

```python
week=['1week', '2week', '3week', '4week', '5week', '6week']
plt.bar(week, mean, width=0.4, color='r')
plt.ylim([0, 10])

plt.xlabel('상영 주차')
plt.ylabel('평균 평점')
plt.title('크루엘라 주차별 평균 평점')
plt.show()
```

# 결론

1. 3편 모두 평균 평점이 상영 1주차에서 가장 높았다.
    - 특히 '파이프라인'의 경우, 나머지 주차와 평균 평점 차이가 가장 컸다.
    
    
2. 3편 모두 평점 개수가 상영 1주차에서 가장 높았다.
    - 특히 '분노의 질주'의 경우, 나머지 주차와 평점 개수 차이가 가장 컸다.
    
    
3. 3편 모두 10점 비율이 상영 1주차가 상영 6주차보다 높았다.
    - 특히 '파이프라인'의 경우, 1주차와 6주차 10점 비율 차이가 가장 컸다.

- 통상적으로 마케팅의 영향으로 개봉 직후 많은 사람들이 관람함. 따라서 평점 개수의 차이는 용납 가능함.  
- 하지만 평균 평점이 주차별로 차이가 컸으며, 특히 개봉 1주차 평균 평점이 가장 높았음.
- 이는 아래 첨부한 기사에서 볼 수 있듯이 매크로를 사용한 조작이 '주로 개봉 첫주 관객 반응을 봐가며 평균 평점 7점을 유지하는 데' 사용되고 있음과 일맥상통함.
- 특히 '최근엔 영화를 관람하고 실망해 가장 낮은 점수인 1점을 집중적으로 다는 관객이 늘어난 탓에 평균 평점 7점을 유지하기 위해 평점 10점을 많이 단다.'라는 영화 마케터의 인터뷰가 앞서 살펴본 1주차와 6주차 10점 비율의 차이의 이유를 설명함.
