```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
```

```python
search_word = input("검색어를 입력해 주세요 : ") # 검색어 입력

options = Options()
# chrome을 전체화면으로 넓히는 옵션입니다.
options.add_argument('--start-fullscreen')

driver = webdriver.Chrome(r'C:\Users\LG\Desktop\2021.1\데이터 언어\12week\chromedriver.exe') 
driver.implicitly_wait(3)

driver.get('https://arxiv.org/')

search = driver.find_element_by_xpath('//*[@id="header"]/div[2]/form/div/div[1]/input') 
# 검색창의 주소 -> f12 -> 검색창 마우스 우클릭 -> 검사 -> copy xpath
search.clear()
search.send_keys(search_word) #검색 단어 받음
time.sleep(1)
search.send_keys(Keys.ENTER) # 검색 엔터 
```

```python
total_header = []
total_author = []
total_date = []


for i in range(1, 11):  ## 10개 크롤링
    head = driver.find_element_by_css_selector(
            f"#main-container > div.content > ol > li:nth-child({i}) > p.title.is-5.mathjax") # 첫 검색창의 copy selector
    author = driver.find_element_by_css_selector(
            f"#main-container > div.content > ol > li:nth-child({i}) > p.authors")
    date = driver.find_element_by_css_selector(
            f"#main-container > div.content > ol > li:nth-child({i}) > p:nth-child(5)")
    
    time.sleep(1)
    total_header.append(head.text)
    total_author.append(author.text)
    total_date.append(date.text)

        
data = {"header": total_header, "author" : total_author, "date": total_date }

df = pd.DataFrame(data, columns = ["header","author","date"])
df
```

```python
df["date"] = df["date"].str.split(';').str[0]
df["date"] = df["date"].str.replace(pat='Submitted ', repl='', regex=True)
df["author"] = df["author"].str.replace(pat='Authors: ', repl='', regex=True)
df
```
