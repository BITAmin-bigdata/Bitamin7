### 영화 평점 비교 : 크리스토퍼 놀란 감독의 영화 네이버 평점을 비교해보자!

```python
import lxml.html
import requests
from bs4 import BeautifulSoup as bs
```

#### 영화 1. 인터스텔라

```python
url = "https://movie.naver.com/movie/bi/mi/point.nhn?code=45290"
res = requests.get(url)
res.status_code                 #상태코드
```

```python
root = lxml.html.fromstring(res.text)
res.text
```

```python
soup = bs(res.text,"lxml")
```

```python
#class=star_score
soup.select("div.star_score")
```

```python
url_sub = "/movie/bi/mi/pointWriteFormList.nhn?code=45290&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false"

#실제 리뷰들이 담긴 장소
url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=45290&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false"
```

```python
res = requests.get(url)
soup = bs(res.text,"lxml")
```

```python
scores_list = soup.select("div.star_score")
for score in scores_list :
    print(score.text.strip())
```

```python
#1~50페이지만큼 평점 가져오기
scorelist = []

for page in range(1,51,1) :
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=45290&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(page)
    res = requests.get(url)
    soup = bs(res.text, "lxml")
    
    scores = soup.select("div.star_score")
    for index in scores :
        print(index.text.strip())
        scorelist.append(index.text.strip())
```

```python
#class=score_reple
reples_list = soup.select("div.score_reple>p")

#관람객
#태그 span, 클래스 ico_viewer
icon = soup.select("span.ico_viewer")
icon

#extract() 함수를 활용해서 관람객 아이콘 삭제
for index in icon :
    index.extract()
```

```python
#1~50페이지만큼 리뷰 가져오기
replelist = []

for page in range(1,51,1) :
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=45290&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="+str(page)
    res = requests.get(url)
    soup = bs(res.text, "lxml")
    
    reples = soup.select("div.score_reple>p")
    icon = soup.select("span.ico_viewer")
    for i in icon :
        i.extract()
    for index in reples :
        print(index.text.strip())
        replelist.append(index.text.strip())
```

```python
#평점과 리뷰 데이터프레임 만들기
import pandas as pd
info = {"interstellar_score" : scorelist, "interstellar_review" : replelist}
interstellar = pd.DataFrame(info)
interstellar
```

## 평점 분석

```python
#interstellar
movie["interstellar_score"].value_counts()
```

```python
#inception
movie["inception_score"].value_counts()
```

```python
#tenet
movie["tenet_score"].value_counts()
```

```python
#the_dark_knight
movie["the_dark_knight_score"].value_counts()
```

```python
movie = movie.astype("int")
```

```python
#평점 평균 비교
print("Interstellar의 평점 평균:", movie["interstellar_score"].mean())
print("Inception의 평점 평균:", movie["inception_score"].mean())
print("Tenet의 평점 평균:", movie["tenet_score"].mean())
print("The Dark Knight의 평점 평균:", movie["the_dark_knight_score"].mean())
```

```python
#subplot 그려서 영화별로 평점 비교하기

import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure()
plt.figure(figsize=(16, 16))

plt.subplot(221)
sns.countplot(x="interstellar_score",data=movie)
plt.title("Interstellar")
plt.ylim(0,500)

plt.subplot(222)
sns.countplot(x="inception_score",data=movie)
plt.title("Inception")
plt.ylim(0,500)

plt.subplot(223)
sns.countplot(x="tenet_score",data=movie)
plt.title("Tenet")
plt.ylim(0,500)

plt.subplot(224)
sns.countplot(x="the_dark_knight_score",data=movie)
plt.title("The_Dark_Knight")
plt.ylim(0,500)

plt.show()
```

2. 평점 분포
