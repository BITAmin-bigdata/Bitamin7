```python
import pandas as pd
import numpy as np
from konlpy.tag import Mecab
```

```python
tokenizer = Mecab()
```

```python
df = pd.read_csv("nmovie.csv", encoding = "utf-8")
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
df.dropna(how = "any", inplace = True)
```

```python
df.query("comment == ''") #다행히 빈 문자열은 없는 모양,,
```

```python
#형태소 추출시 데이터프레임 인덱스를 활용할 예정이므로 인덱스 재정렬
df.reset_index(drop = True, inplace = True)
df
```

tokenizer.pos 이용시 튜플 형태로 각 문장의 형태소와 품사 출력

이 중 품사가 체언(N으로 시작), 용언(V으로 시작), 부사/관형사(M으로 시작)만 추출

```python
tokenizer.pos(df["comment"][1])[:10] #예시
```

```python
df_pp = df.copy()
for i, com in enumerate(df_pp["comment"]):
    clean_words = []
    for word in tokenizer.pos(com): #코멘트 한 줄에 대한 형태소와 품사 튜플
        if word[1][0] in ["N", "V", "M"]: #튜플 내 품사(2번째) 첫 글자
            clean_words.append(word[0]) #튜플 내 형태소(1번째)
    doc = " ".join(clean_words)
    print(doc)
    df_pp["comment"][i] = doc
print(df_pp)
```

```python
print(df_pp.isnull().sum()) #결측치 확인
drop_idx = df_pp.query("comment == ''").index
print(drop_idx) #빈 문자열 확인 및 인덱스 추출
```

```python
df_final = df_pp.drop(drop_idx, axis = 0).reset_index(drop = True)
df_final
```

```python
#rating 분포 확인 > 상당한 불균형 데이터
(df_final["rating"].value_counts() / df_final.shape[0]) * 100
```

위의 불균형 해소 방안
1. rating값을 긍정/부정의 binary output으로 재정립(cutoff threshold 설정 중요)
2. 크롤링 시 rating값에 조건을 걸어 처음부터 일정 비율로 맞추기

```python
df_final.to_csv("nmovie_f.csv", encoding = "utf-8")
```
