## 네이버 뉴스에서 기사들의 댓글수, 감정, 제목을 추출한 뒤 데이터프레임으로 만들려고 합니다..!

```python
from selenium.webdriver import Chrome
from time import sleep
import lxml.html
import requests
```

```python
comment_list =[]
recommend_list =[]
good_list =[]
impress_list=[]
angry_list=[]
sad_list=[]
title_list =[]
#댓글 수, 각 공감 수, 제목들을 리스트 형태로 만든 뒤에 dataframe형태로 만들어줄 예정입니다.
```

```python
def mood(): #한 기사에서 모든 댓글,공감 수,제목들을 추출하는 함수 생성
    try:
        comment = browser.find_elements_by_css_selector('#alex-header > em') #댓글 개수
        sleep(0.01)
        recommend = browser.find_elements_by_css_selector('#mArticle > div.foot_view > div.emotion_wrap > div.emotion_list > div > div > div > div.selectionbox.type-RECOMMEND.unselected > span.count') #추천해요 개수
        sleep(0.01)
        good = browser.find_elements_by_css_selector('#mArticle > div.foot_view > div.emotion_wrap > div.emotion_list > div > div > div > div.selectionbox.type-LIKE.unselected > span.count') #좋아요 개수
        sleep(0.01)
        impress = browser.find_elements_by_css_selector('#mArticle > div.foot_view > div.emotion_wrap > div.emotion_list > div > div > div > div.selectionbox.type-IMPRESS.unselected > span.count') #감동이에요 개수
        sleep(0.01)
        angry = browser.find_elements_by_css_selector('#mArticle > div.foot_view > div.emotion_wrap > div.emotion_list > div > div > div > div.selectionbox.type-ANGRY.unselected > span.count') #화나요 개수
        sleep(0.01)
        sad = browser.find_elements_by_css_selector('#mArticle > div.foot_view > div.emotion_wrap > div.emotion_list > div > div > div > div.selectionbox.type-SAD.unselected > span.count') #슬퍼요 개수
        sleep(0.01)
        title = browser.find_elements_by_css_selector('#cSub > div > h3') #제목
        sleep(0.01)
        
        comment_list.append(comment[0].text)
        recommend_list.append(recommend[0].text)
        good_list.append(good[0].text)
        impress_list.append(impress[0].text)
        angry_list.append(angry[0].text)
        sad_list.append(sad[0].text)
        title_list.append(title[0].text)
    except:
        None
```

```python
browser = Chrome()
```

```python
url = 'https://news.daum.net/breakingnews/politics'
```

```python
url_of_page=[]
def get_url_of_page():
    for i in range(1,31):
        pages = [f'https://news.daum.net/breakingnews/politics?page={i}'] #1페이지부터 30페이지까지 페이지 지정
        for page in pages: 
            res = requests.get(page)
            root = lxml.html.fromstring(res.text);root
            links = root.cssselect('.tit_thumb a') 
            for link in links:
                url_of_page.append(link.attrib['href']) #각 page에 있는 기사들의 url을 모두 url_of_page에 넣어줌
```

```python
get_url_of_page()
url_of_page 
```

```python
def mooddata(url): #각 url에서 mood()함수 실행
    browser.get(url.format(4))
    sleep(2)
    mood()
    sleep(0.5)
```

```python
for url in url_of_page: #위의 mooddata()를 반복해서 실행
    mooddata(url)
```

```python
import pandas as pd 
df = pd.DataFrame({'댓글 수': comment_list,'추천해요':recommend_list,'좋아요':good_list,'감동이에요':impress_list,'화나요':angry_list,'슬퍼요':sad_list,'제목':title_list,'URL':url})
df
```
